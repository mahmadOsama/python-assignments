# EX 1 - Distribution Plots

Create a histogram of a given dataset using `histplot`.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Titanic-Dataset.csv')
import pandas as p
df.head()

df.columns

print(df.describe())

print(df.isnull().sum())
```

```python
#before draw the histogram we need to clean data from NANS values and missing values
#i Here I preferred to use the mean instead of giving zero values ​​so as not to create a gap in the data.
df["Age"] = df["Age"].fillna(df["Age"].median())
#Fill missing values ​​in Embarked with the most frequent value
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
#Since it has many missing values, I preferred to drop it.
df = df.drop(columns=["Cabin"])


print(df.isnull().sum())

```

```python
sns.histplot(data=df, x="Age", bins=30, kde=True)
plt.title("Distribution of Passenger Ages on Titanic")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

Generate a KDE plot of the same dataset using `kdeplot`.

```python

sns.kdeplot(data=df, x="Fare", fill=True)

plt.title("Kernel Density Estimate of Passenger Fares on Titanic")
plt.xlabel("Fare")
plt.ylabel("Density")
plt.show()
```

Use `jointplot` to display the relationship between two numerical variables.

```python



sns.jointplot(data=df, x="Age", y="Fare", kind="scatter")

plt.show()
```

```python
#Converting makes it easier to color and classify correctly while drawing. So I converted it here to category.
df['Pclass'] = df['Pclass'].astype('category')
df['Sex'] = df['Sex'].astype('category')
df['Embarked'] = df['Embarked'].astype('category')
```

Create a `pairplot` to visualize pairwise relationships in a dataset.

```python
sns.pairplot(df, hue='Survived', vars=['Age', 'SibSp', 'Parch', 'Fare'])
plt.show()
```


## EX 2 - Categorical Data Plots
Create a `barplot` to show the average values of a categorical variable.

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Pclass', y='Age', data=df)
plt.title("Average Age per Pclass")
plt.xlabel("Pclass (Passenger Class)")
plt.ylabel("Average Age")
plt.show()
```

Generate a `countplot` to show the frequency of different categories.

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Pclass', data=df)
plt.title("Number of Passengers per Pclass")
plt.xlabel("Pclass (Passenger Class)")
plt.ylabel("Count")
plt.show()
```

Use `boxplot` to visualize the distribution of numerical data across categories.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title("Age Distribution by Pclass")
plt.xlabel("Pclass (Passenger Class)")
plt.ylabel("Age")
plt.show()
```

Create a `violinplot` to combine KDE and boxplot visualizations.

```python
plt.figure(figsize=(8,5))
sns.violinplot(x='Pclass', y='Age', data=df, inner='box', palette='Pastel1')
plt.title("Age Distribution by Pclass")
plt.xlabel("Pclass (Passenger Class)")
plt.ylabel("Age")
plt.show()
```

Display individual data points in a category using `stripplot`.

```python
plt.figure(figsize=(8,5))
sns.stripplot(x='Pclass', y='Age', data=df, jitter=True, palette='Set2')
plt.title("Individual Age Data Points by Pclass")
plt.xlabel("Pclass (Passenger Class)")
plt.ylabel("Age")
plt.show()
```

Use `swarmplot` to create a more readable version of `stripplot`.

```python
plt.figure(figsize=(8,5))
sns.swarmplot(x='Pclass', y='Age', data=df, palette='Set2', size=4)
plt.title("Individual Age Data Points by Pclass (Swarmplot)")
plt.xlabel("Pclass (Passenger Class)")
plt.ylabel("Age")
plt.show()
```

# EX 3 - Matrix Plots
Create a heatmap to display correlations in a dataset using `heatmap`.

```python
 #Each numeric column in the data represents a series of numbers ( Age, Fare, SibSp, Parch).

# The correlation matrix calculates the Pearson correlation coefficient for each pair of these columns.
# The value ranges from -1 to 1:
# 1 → Perfect positive correlation (as one increases, the other increases by approximately the same proportion).
# -1 → Perfect negative correlation (as one increases, the other decreases by approximately the same proportion).
# 0 → No clear linear correlation.

# The purpose of this is to:
# Understand which columns are correlated with each other, especially with the target column, such as Survived.
numeric_cols = df.select_dtypes(include='number')
corr_matrix = numeric_cols.corr()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Titanic Numeric Columns")
plt.show()
```

Use `clustermap` to cluster data based on similarity.

```python
sns.clustermap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Clustermap of Titanic Numeric Columns", pad=80)  
plt.show()
```

# EX 4 - Grids
Implement a `PairGrid` to customize pairwise plots.

```python
umeric_cols = ['Age', 'Fare', 'SibSp', 'Parch']

g = sns.PairGrid(df, vars=numeric_cols, hue='Survived', palette='Set1')

g.map_upper(sns.scatterplot)     
g.map_lower(sns.kdeplot, fill=True, warn_singular=False)  
g.map_diag(sns.histplot, kde=True) 

g.add_legend(title='Survived')

plt.show()
```

Use `FacetGrid` to create multiple subplots based on a categorical variable.

```python
g = sns.FacetGrid(df, col='Pclass', hue='Sex', palette='Set1', height=4, aspect=1)
g.map(sns.histplot, 'Age', kde=True, bins=20, alpha=0.5) 


g.add_legend(title='Sex')

g.set_axis_labels("Age", "Count")
g.set_titles("Pclass {col_name}")

plt.show()
```

Implement a `JointGrid` to create a customized joint distribution plot.

```python

g = sns.JointGrid(data=df, x='Age', y='Fare', height=6)

g.plot(sns.scatterplot, sns.histplot)


g.ax_marg_x.hist(df['Age'], bins=20, color='skyblue', alpha=0.6)
g.ax_marg_y.hist(df['Fare'], bins=20, orientation='horizontal', color='salmon', alpha=0.6)


plt.suptitle("Joint Distribution of Age and Fare", y=1.02, fontsize=16)

plt.show()
```
