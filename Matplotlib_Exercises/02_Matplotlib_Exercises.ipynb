** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid1.inset_locator import inset_axes


x = np.linspace(0, 100, 500)
y = x ** 2  

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, y, color='blue')
ax.set_xlabel('X')
ax.set_ylabel('Z')
ax.set_xlim(0, 100)
ax.set_ylim(0, 10000)


ax_inset = inset_axes(ax, width="40%", height="40%", loc='upper center')


x_inset = np.linspace(20, 22, 100)
y_inset = np.linspace(40, 45, 100)  

ax_inset.plot(x_inset, y_inset, color='blue')
ax_inset.set_xlim(20, 22)
ax_inset.set_ylim(30, 50)
ax_inset.set_title("Inset Plot")
ax_inset.set_xlabel("X")
ax_inset.set_ylabel("Y")

plt.tight_layout()
plt.show()
```

```python

from mpl_toolkits.axes_grid1.inset_locator import inset_axes
x = np.linspace(0, 100, 500)
y = x ** 2

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 10))


ax1.plot(x, y, color='blue')
ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax1.set_xlim(0, 100)
ax1.set_ylim(0, 10000)

ax1_inset = inset_axes(ax1, width="40%", height="40%", loc='upper center')
x_inset = np.linspace(20, 22, 100)
y_inset = np.linspace(40, 45, 100)
ax1_inset.plot(x_inset, y_inset, color='blue')
ax1_inset.set_xlim(20, 22)
ax1_inset.set_ylim(30, 50)
ax1_inset.set_title("zoom")
ax1_inset.set_xlabel("X")
ax1_inset.set_ylabel("Y")


ax2.plot(x, y, color='blue')
ax2.set_xlabel('X')
ax2.set_ylabel('Z')
ax2.set_xlim(0, 100)
ax2.set_ylim(0, 10000)


ax2_inset = inset_axes(ax2, width="40%", height="40%", loc='upper center')
ax2_inset.plot(x_inset, y_inset, color='blue')
ax2_inset.set_xlim(20, 22)
ax2_inset.set_ylim(30, 50)
ax2_inset.set_title("zoom")
ax2_inset.set_xlabel("X")
ax2_inset.set_ylabel("Y")

plt.tight_layout()
plt.show()
```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python


x = np.arange(0, 101)  
y = 2 * x              
z = x ** 2            


fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))


axes[0].plot(x, y, 'b--')  
axes[0].set_title("X vs Y Plot (Dashed)")
axes[0].set_xlabel("X")
axes[0].set_ylabel("Y")

axes[1].plot(x, z, 'r:')  
axes[1].set_title("X vs Z Plot (Dotted)")
axes[1].set_xlabel("X")
axes[1].set_ylabel("Z")


plt.tight_layout()
plt.show()
```
