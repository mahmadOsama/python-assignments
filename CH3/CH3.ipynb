When you write multiple f-strings one after another without printing them separately, only the last expression will actually be evaluated and shown if you just run the cell or the script without using print(). For example, in:

```python

age = 25
f"My age is {age}"
# f"Next year I will be {age + 1} years old"

```

## Indexing

* Strings are sequences of characters. Each character in a string has an index according to its position in the sequence.

* The string "ABCD" has for characters, each with the following index:

| Char | Index |
|---|---|
|A|0|
|B|1|
|C|2|
|D|3|

We can access a specific character in a string using its index.

---



Types of index:

1) **Positive indexing** starts from 0 for the first item, 1 for the second item, and so on.

2) **Negative indexing** starts from -1 for the last item, -2 for the second last, and so on.

```python
#    0 b, 1 a, 2 t, 3 o, 4 o, 5 l
x = "batool"
print(x[2])
print(x[-4])
```

## Slicing

- Slicing is a technique used to extract a portion of a string (or any sequence like list or tuples) by specifying a range of indices.

- Slicing creates a new output containing the specified elements.

---



- Slicing Elements:

1. start: The index to begin the slice.
2. end: The index to end the slice.
3. step: The interval between elements.
